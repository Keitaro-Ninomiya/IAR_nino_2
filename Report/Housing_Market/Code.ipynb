```julia
def Adj(Value,Type):
    if Type==1:
        if np.sign(Value)==-1:
            return("negative")
        if np.sign(Value)==1:
            return("positive")
        else:
            return('unchanged')
    if Type==2:
        if np.sign(Value)==-1:
            return("down")
        if np.sign(Value)==1:
            return("up")
        else:
            return('unchanged at')
    if Type==3:
        if np.sign(Value)==-1:
            return("less")
        if np.sign(Value)==1:
            return("more")
        else:
            return('unchanged at')
        
    if Type==4:
        if np.sign(Value)==-1:
            return("decrease")
        if np.sign(Value)==1:
            return("increase")
        else:
            return('unchanged')

def Comma(Value):
    return(locale.format_string("%d", Value,-2))
```
